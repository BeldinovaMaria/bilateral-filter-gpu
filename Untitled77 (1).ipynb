```python
# Создаем CUDA файл с текстурной памятью (рабочий синтаксис для CUDA 12+)
cuda_file_content = """
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <cuda_runtime.h>

// Текстурная память - объявление texture reference
texture<float, 2, cudaReadModeElementType> texRef;

// Ядро с использованием текстурной памяти
__global__ void bilateralFilterKernel(float* output, int width, int height,
                                       float sigma_d, float sigma_r) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    if (x >= width || y >= height) return;

    float spatial_weights[9];
    float sigma_d_sq = 2 * sigma_d * sigma_d;

    int dx[] = {-1, 0, 1, -1, 0, 1, -1, 0, 1};
    int dy[] = {-1, -1, -1, 0, 0, 0, 1, 1, 1};

    // Пространственные веса
    for (int k = 0; k < 9; k++) {
        float dist_sq = dx[k] * dx[k] + dy[k] * dy[k];
        spatial_weights[k] = expf(-dist_sq / sigma_d_sq);
    }

    // Доступ к центральному пикселю через текстурную память
    float center_val = tex2D(texRef, (float)x + 0.5f, (float)y + 0.5f);

    float sum_weights = 0.0f;
    float sum_weighted_pixels = 0.0f;

    for (int k = 0; k < 9; k++) {
        int nx = x + dx[k];
        int ny = y + dy[k];

        nx = max(0, min(nx, width - 1));
        ny = max(0, min(ny, height - 1));

        // Доступ к соседним пикселям через текстурную память
        float neighbor_val = tex2D(texRef, (float)nx + 0.5f, (float)ny + 0.5f);

        float intensity_diff = neighbor_val - center_val;
        float range_weight = expf(-(intensity_diff * intensity_diff) / (2 * sigma_r * sigma_r));

        float weight = spatial_weights[k] * range_weight;
        sum_weights += weight;
        sum_weighted_pixels += weight * neighbor_val;
    }

    output[y * width + x] = sum_weighted_pixels / sum_weights;
}

// Функция для привязки текстурной памяти
void bindTexture(float* d_data, int width, int height) {
    cudaChannelFormatDesc channelDesc = cudaCreateChannelDesc<float>();
    cudaBindTexture2D(NULL, &texRef, d_data, channelDesc, width, height, width * sizeof(float));

    // Настройка текстурной памяти
    texRef.addressMode[0] = cudaAddressModeClamp;
    texRef.addressMode[1] = cudaAddressModeClamp;
    texRef.filterMode = cudaFilterModePoint;
    texRef.normalized = false;
}

// Функция для отвязки текстуры
void unbindTexture() {
    cudaUnbindTexture(texRef);
}
"""

# Сохраняем в файл
with open('bilateral_texture.cu', 'w') as f:
    f.write(cuda_file_content)

print(" Файл bilateral_texture.cu создан")
```
